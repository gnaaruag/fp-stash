```python
import pandas as pd 
import numpy as np
```

```python
dataset=pd.read_csv(r"C:\Users\arunm\Downloads\Debernardi et al 2020 data.csv")
dataset.diagnosis = dataset.diagnosis == 3
dataset
```

```python
print(dataset.patient_cohort.value_counts())
print(dataset.sample_origin.value_counts())
```

```python
dataset.drop(columns=["sample_id"],inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
l=list(dataset.columns)
l.insert(0,'patient_cohort')
x_1=pd.DataFrame(ct.fit_transform(dataset),columns=l)
x_1
```

```python
print(x_1.sample_origin.value_counts())
print(x_1.stage.isna().value_counts())
x_1.benign_sample_diagnosis.value_counts()
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='constant',fill_value='0')
a=si.fit_transform(x_1[["stage"]])
l=[]
for i in range(len(a)):
    l.append(a[i][0])
x_1["stage"]=l 
x_1["stage"]
```

```python
si_1=SimpleImputer(strategy='constant',fill_value='0')
a=si_1.fit_transform(x_1[["benign_sample_diagnosis"]])
l=[]
for i in range(len(a)):
    l.append(a[i][0])
x_1["benign_sample_diagnosis"]=l
```

```python
si_2=SimpleImputer(strategy='constant',fill_value=0)
a=si_2.fit_transform(x_1[["REG1A"]])
l=[]
for i in range(len(a)):
    l.append(a[i][0])
x_1["REG1A"]=l  
x_1.plasma_CA19_9.isna().value_counts()
```

```python
si_3=SimpleImputer(strategy='constant',fill_value=0)
a=si_3.fit_transform(x_1[["plasma_CA19_9"]])
l=[]
for i in range(len(a)):
    l.append(a[i][0])
x_1["plasma_CA19_9"]=l  
x_1["plasma_CA19_9"]
```

```python
print(x_1.TFF1.isna().value_counts())
print(x_1.stage.value_counts())
order_stage=["0","IA","IB","I","IIA","IIB","II","III","IV"]
x_1["stage"]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[order_stage])
x_1["stage"]=oe.fit_transform(x_1[["stage"]])
x_1["stage"].value_counts()
```

```python
x_1.benign_sample_diagnosis.value_counts()
```

```python
for i in range(len(x_1["benign_sample_diagnosis"])):
    if (x_1["benign_sample_diagnosis"][i]=='0'):
        x_1["benign_sample_diagnosis"][i]=0
    else:
        x_1["benign_sample_diagnosis"][i]=1
```

```python

x_1.to_csv('output.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
x = x_1.drop('diagnosis', axis=1)
y = x_1['diagnosis']

class_counts = pd.Series(y).value_counts()
print(class_counts)

# Split the data into train and test sets

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.to_csv('x_train.csv', index=False)
x_test.to_csv('x_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```
